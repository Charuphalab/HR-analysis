```python
print("Existing columns:", df.columns)
```

```python
columns_to_drop = ['Age', 'EmployeeNumber', 'StandardHours', 'EmployeeCount']

# Check if each column exists before attempting to drop it
for column in columns_to_drop:
    if column in df.columns:
        df = df.drop(column, axis=1)
    else:
        print(f"Column '{column}' not found in the DataFrame")

# Now, df contains the DataFrame without the specified columns (if they existed)
```
